```python
input_file = './cataciones_original.xlsx'
dfc = pd.read_excel(input_file, sheet_name=None)
dfc.head(2)
```

```python
dr = dfc.filter(items=['OL_TOT','REGIÓN'])
dr.head()
```

```python
pd.to_numeric(dr.OL_TOT, errors='coerce')
```

```python
RegionCounts={}
i =0
dr['C'] = 0
for i in range(len(dr.OL_TOT)):
    dr.C[i]=i

print(dr)
```

```python
dx = dr
```

```python
dx.dropna(inplace=True)
dx
```

```python
i=0
for i in dx.C:
    if(dx.REGIÓN[i] !=0):
        if(dx.OL_TOT[i] >= 81):
            if dx.REGIÓN[i] in RegionCounts:
                RegionCounts[dx.REGIÓN[i]] = RegionCounts[dx.REGIÓN[i]] + 1
            else:
                RegionCounts[dx.REGIÓN[i]] = 1

    
print(RegionCounts)

```
