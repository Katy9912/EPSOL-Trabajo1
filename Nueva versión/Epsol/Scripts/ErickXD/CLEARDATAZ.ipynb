```python
import csv
import pandas as pd
import os
```

```python
def limpiar(path):
    data = pd.read_csv(path, header= None, sep = '\n' )
    data = data[0].str.split(',', expand=True)
    index=int(data[data[0]=="Record"].index[0])
    data.drop(range(0,index),axis=0,inplace=True)
    data.reset_index(drop=True,inplace=True)
    data.columns=data.iloc[0]
    data.drop(0,axis=0,inplace=True)
    data.dropna(axis = 1, how ='all',inplace=True)
    return data.copy()


```

```python
LPDS_Dir = "./"
filename = "LDP Export 2020-may.-22_01-00-24_Recorder2.csv"
path = os.path.join(LPDS_Dir,filename)

LPDS= limpiar(path)
LPDS.head(1000)
```
